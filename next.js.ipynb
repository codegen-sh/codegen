Based on our [Slack conversation](https://thecodegen.slack.com/archives/C07HXT5NN75/p1739994947429139?thread_ts=1738268741.592289&cid=C07HXT5NN75), this will flag + link all imports that:
- import *from* `DIR_NAME` directly
- the import is *outside* of DIRNAME

This uses the [Files and Directories](https://docs.codegen.com/building-with-codegen/files-and-directories#files-and-directories) APIs, as well as [Imports](https://docs.codegen.com/building-with-codegen/imports#the-import-api)

```python
from codegen.sdk.core.detached_symbols.code_block import CodeBlock


def is_env_conditional(statement):
    if not isinstance(statement, IfBlockStatement):
        return False
    return "process.env" in statement.condition.source


def get_all_statements(block: CodeBlock):
    statements = []
    for statement in block.statements:
        statements.append(statement)
        if hasattr(statement, "code_block"):
            statements += get_all_statements(statement.code_block)
    return statements
```

```python
DIR_NAME = "packages/next/src/"
directory = codebase.get_directory(DIR_NAME)

for file in directory.files:
    env_conditionals = [s for s in get_all_statements(file.code_block) if is_env_conditional(s)]
    for c in env_conditionals:
        print(c.github_url)
```

```python
from codegen.sdk.core.statements.if_block_statement import IfBlockStatement

DIR_NAME = "packages/next/src/client/components/react-dev-overlay"
directory = codebase.get_directory(DIR_NAME)


def is_env_conditional(statement):
    if not isinstance(statement, IfBlockStatement):
        return False
    return "process.env" in statement.condition.source


for file in directory.files:
    for imp in file.inbound_imports:
        if imp.file not in directory:
            # Grab process.env statements
            file = imp.to_file
            statements = get_all_statements(file.code_block.statements)
            conditionals = [x for x in statements if is_env_conditional(x)]

            # See if the import is beneath one of the process.env statementss
            if not any(imp.source in c.source for c in conditionals):
                continue

            print(f"❌ Detected bad import:")
            print(f"  ⚛️ File: {imp.file.filepath}")
            print(f"  ⬇️ Imported symbol: {imp.name}")
            print(f"  🔗 Link: {imp.github_url}")
```
